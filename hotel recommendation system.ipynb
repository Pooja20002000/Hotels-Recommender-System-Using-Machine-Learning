```python
reviwers = pd.read_csv('Datas/reviwer.csv', sep=",", encoding='latin-1')
```

```python
hotels_with_reviwer = hotels.merge(reviwers, on="Hotel_Name")
```

```python
hotels_with_reviwer.shape
```

```python
rate_hotel = hotels_with_reviwer.groupby('Hotel_Name')['Reviewer_Score'].count().reset_index()
```

```python
rate_hotel
```

```python
rate_hotel.head()
```

```python
final = rate_hotel[rate_hotel['Reviewer_Score']>1000]
```

```python
final.head()
```

```python
final.sample()
```

```python
final.drop_duplicates(['Hotel_Name','Reviewer_Score'], inplace=True)
```

```python
final.shape
```

```python
hotel_pivot = hotels.pivot_table(columns='Reviewer_Nationality', index='Hotel_Name', values='Reviewer_Score')
```

```python
hotel_pivot
```

```python
hotel_pivot.fillna(0, inplace=True)
```

```python
hotel_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
hotel_sparse = csr_matrix(hotel_pivot)
```

```python
hotel_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(hotel_sparse)
```

```python
distance, suggestion = model.kneighbors(hotel_pivot.iloc[1,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(hotel_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl','wb'))
pickle.dump(hotels_name, open('artifacts/hotels_name.pkl','wb'))
pickle.dump(final, open('artifacts/final.pkl','wb'))
pickle.dump(hotel_pivot, open('artifacts/hotel_pivot.pkl','wb'))
```

```python
def recommend_hotel(hotel_name):
    hotel_id = np.where(hotel_pivot.index == hotel_name)[0][0]
    distance, suggestion = model.kneighbors(hotel_pivot.iloc[hotel_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range (len(suggestion)):
        hotels = hotel_pivot.index[suggestion[i]]
        for j in hotels:
            print(j)
```

```python
hotel_name = 'Hotel Panache'
recommend_hotel(hotel_name)
```
